```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
df['Price_per_sqft'] = df['Price'] / df['Area']

amenities = [
    'Resale', 'MaintenanceStaff', 'Gymnasium', 'SwimmingPool', 'LandscapedGardens',
    'JoggingTrack', 'RainWaterHarvesting', 'IndoorGames', 'ShoppingMall', 'Intercom',
    'SportsFacility', 'ATM', 'ClubHouse', 'School', '24X7Security', 'PowerBackup',
    'CarParking', 'StaffQuarter', 'Cafeteria', 'GolfCourse', 'LiftAvailable',
    'VaastuCompliant', 'Microwave', 'TV', 'DiningTable', 'Sofa', 'Wardrobe', 'Refrigerator'
]
df['Total_Amenities'] = df[amenities].sum(axis=1)

features = ['Price_per_sqft', 'Area', 'No. of Bedrooms', 'Total_Amenities']
X = df[features]
y = df['Price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_preds)
lr_r2 = r2_score(y_test, lr_preds)

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_preds)
rf_r2 = r2_score(y_test, rf_preds)

print("Linear Regression:")
print(f"  MSE: {lr_mse:.2f}")
print(f"  R2 Score: {lr_r2:.2f}")

print("\nRandom Forest:")
print(f"  MSE: {rf_mse:.2f}")
print(f"  R2 Score: {rf_r2:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(y_test, lr_preds)

rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_preds)

print("Linear Regression - Evaluation Metrics:")
print(f"  MSE: {lr_mse:.2f}")
print(f"  RMSE: {lr_rmse:.2f}")
print(f"  R² Score: {lr_r2:.2f}")

print("\nRandom Forest - Evaluation Metrics:")
print(f"  MSE: {rf_mse:.2f}")
print(f"  RMSE: {rf_rmse:.2f}")
print(f"  R² Score: {rf_r2:.2f}")
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import joblib

df = pd.read_csv('merged_files.csv')

df['Price_per_sqft'] = df['Price'] / df['Area']

amenities = [
    'Resale', 'MaintenanceStaff', 'Gymnasium', 'SwimmingPool', 'LandscapedGardens',
    'JoggingTrack', 'RainWaterHarvesting', 'IndoorGames', 'ShoppingMall', 'Intercom',
    'SportsFacility', 'ATM', 'ClubHouse', 'School', '24X7Security', 'PowerBackup',
    'CarParking', 'StaffQuarter', 'Cafeteria', 'GolfCourse', 'LiftAvailable',
    'VaastuCompliant', 'Microwave', 'TV', 'DiningTable', 'Sofa', 'Wardrobe', 'Refrigerator'
]
df['Total_Amenities'] = df[amenities].sum(axis=1)
features = ['Price_per_sqft', 'Area', 'No. of Bedrooms', 'Total_Amenities']
X = df[features]
y = df['Price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)

xgb_preds = xgb_model.predict(X_test)

xgb_mse = mean_squared_error(y_test, xgb_preds)
xgb_rmse = np.sqrt(xgb_mse)
xgb_r2 = r2_score(y_test, xgb_preds)

print("XGBoost Model - Evaluation Metrics:")
print(f"  MSE: {xgb_mse:.2f}")
print(f"  RMSE: {xgb_rmse:.2f}")
print(f"  R² Score: {xgb_r2:.2f}")

joblib.dump(xgb_model, 'xgboost_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
import streamlit as st
import joblib
import numpy as np

# Load the model and scaler
xgb_model = joblib.load('xgboost_model.pkl')
scaler = joblib.load('scaler.pkl')

# Streamlit App
st.title('House Price Prediction with XGBoost')

# Input fields for user input
area = st.number_input('Area (sq. ft.)', min_value=1, max_value=10000, step=1)
bedrooms = st.number_input('Number of Bedrooms', min_value=1, max_value=10, step=1)
amenities = st.slider('Total Amenities', min_value=0, max_value=30, step=1)
price_per_sqft = st.number_input('Price per Sqft', min_value=1, max_value=10000, step=1)

# Prediction on button click
if st.button('Predict Price'):
    input_data = np.array([[price_per_sqft, area, bedrooms, amenities]])
    input_scaled = scaler.transform(input_data)

    # Make prediction
    prediction = xgb_model.predict(input_scaled)

    st.write(f"Predicted House Price: ${prediction[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
def plot_residuals(y_test, lr_preds, rf_preds):
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    sns.residplot(x=lr_preds, y=y_test - lr_preds, lowess=True, line_kws={'color': 'red'})
    plt.title('Linear Regression Residuals')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')

    plt.subplot(1, 2, 2)
    sns.residplot(x=rf_preds, y=y_test - rf_preds, lowess=True, line_kws={'color': 'blue'})
    plt.title('Random Forest Residuals')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')

    plt.tight_layout()
    plt.show()

plot_residuals(y_test, lr_preds, rf_preds)
```

```python
# Plotting Actual vs Predicted values
def plot_actual_vs_predicted(y_test, lr_preds, rf_preds):
    plt.figure(figsize=(14, 6))

    # Linear Regression Actual vs Predicted
    plt.subplot(1, 2, 1)
    plt.scatter(y_test, lr_preds, color='red')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', linewidth=2)
    plt.title('Linear Regression: Actual vs Predicted')
    plt.xlabel('Actual Prices')
    plt.ylabel('Predicted Prices')

    # Random Forest Actual vs Predicted
    plt.subplot(1, 2, 2)
    plt.scatter(y_test, rf_preds, color='green')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', linewidth=2)
    plt.title('Random Forest: Actual vs Predicted')
    plt.xlabel('Actual Prices')
    plt.ylabel('Predicted Prices')

    plt.tight_layout()
    plt.show()

# Example: plot actual vs predicted
plot_actual_vs_predicted(y_test, lr_preds, rf_preds)
```

```python
import gradio as gr
import joblib
import numpy as np

xgb_model = joblib.load('xgboost_model.pkl')
scaler = joblib.load('scaler.pkl')

def predict(area, bedrooms, amenities, price_per_sqft):
    input_data = np.array([[price_per_sqft, area, bedrooms, amenities]])

    input_scaled = scaler.transform(input_data)

    prediction = xgb_model.predict(input_scaled)[0]

    return f"Predicted House Price: Rs.{prediction:,.2f}"

interface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Textbox(label="Area (sq. ft.)", value="1000"),
        gr.Textbox(label="Number of Bedrooms", value="3"),
        gr.Slider(minimum=0, maximum=30, step=1, label="Total Amenities", value=5),
        gr.Textbox(label="Price per Sqft", value="2500")
    ],
    outputs="text"
)

interface.launch()
```
